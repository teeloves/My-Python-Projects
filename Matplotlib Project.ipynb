```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df= pd.read_csv('/Users/mac/Downloads/london_weather.csv')
df.info()
```

```python
df['date'] = df['date'].astype('str')
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'])
df.info()
```

```python
df.head()
```

```python
df.set_index('date', inplace = True)
df.head()
```

```python
annualData = df.resample('YE').mean()
annualData.head()
```

```python
x = annualData.index
y = annualData['cloud_cover']
plt.figure(
    figsize = (20, 7)
)
plt.plot(x, y)
plt.show()
```

```python
plt.figure(
    figsize = (20, 15)
)

x = annualData.index
y = annualData['sunshine']
plt.subplot(3, 2, 1)
plt.title('Annual Sunshine In London')
plt.plot(x, y)

x = annualData.index
y = annualData['cloud_cover']
plt.subplot(3, 2, 2)
plt.title('Annual Cloud Cover In London')
plt.plot(x, y)

x = annualData.index
y = annualData['global_radiation']
plt.subplot(3, 2, 3)
plt.title('Annual Radiation In London')
plt.plot(x, y)

x = annualData.index
y = annualData['precipitation']
plt.subplot(3, 2, 4)
plt.title('Annual Precipitation In London')
plt.plot(x, y)

x = annualData.index
y = annualData['pressure']
plt.subplot(3, 2, 5)
plt.title('Annual Pressure In London')
plt.plot(x, y)

x = annualData.index
y = annualData['snow_depth']
plt.subplot(3, 2, 6)
plt.title('Annual Snow Depth In London')
plt.plot(x, y)

plt.show()
```
